```python
import matplotlib.pyplot as plt

# Given data
matrix_sizes = [5, 10, 50, 100, 500, 1000, 2000]
execution_times = [0.000000, 0.000000, 0.000000, 0.010000, 0.540000, 3.810000, 35.600000]

# Define a color map
colors = range(len(matrix_sizes))

# Plotting the data with different colors
plt.scatter(matrix_sizes, execution_times, c=colors, cmap='viridis', marker='o')

# Adding a colorbar for reference
plt.colorbar(label='Data Points')

# Adding labels and title
plt.xlabel('Matrix Size')
plt.ylabel('Execution Time (s)')
plt.title('Execution Time and Martix Size')

# Display the plot
plt.show()
```
