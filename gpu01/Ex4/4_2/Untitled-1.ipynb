```python
#Plots for one Block Shared, REgister, GLobal

fig, ax = plt.subplots(4,figsize=(15, 28))

x_steps = 32

for size,group in list_of_df[0].groupby('size'):
    ax[0].plot(group['bDim'], group['bw'], label=f"Size {size}")

for size,group in list_of_df[1].groupby('size'):
    ax[1].plot(group['bDim'], group['bw'], label=f"Size {size}")

for size,group in list_of_df[2].groupby('size'):
    ax[2].plot(group['bDim'], group['bw'], label=f"Size {size}")
    

for size,group in list_of_df[3].groupby('size'):
    ax[3].plot(group['bDim'], group['bw'], label=f"Size {size}")
    



ax[0].set_xlabel('Threads per Block')
ax[0].set_ylabel('Bandwidth in GB/s')
ax[0].legend(title='Size transfered')
ax[0].grid(True)
ax[0].set_title('Global2Shared')
ax[0].set_xticks(np.arange(0,1025, x_steps))

ax[1].set_xlabel('Threads per Block')
ax[1].set_ylabel('Bandwidth in GB/s')
ax[1].legend(title='Size transfered')
ax[1].grid(True)
ax[1].set_title('Shared2Global')
ax[1].set_xticks(np.arange(0,1025, x_steps))

ax[2].set_xlabel('Threads per Block')
ax[2].set_ylabel('Bandwidth in GB/s')
ax[2].legend(title='Size transfered')
ax[2].grid(True)
ax[2].set_title('Shared2Register')
ax[2].set_xticks(np.arange(0,1025, x_steps))

ax[3].set_xlabel('Threads per Block')
ax[3].set_ylabel('Bandwidth in GB/s')
ax[3].legend(title='Size transfered')
ax[3].grid(True)
ax[3].set_title('Register2Shared')
ax[3].set_xticks(np.arange(0,1025, x_steps))

plt.show()
```
